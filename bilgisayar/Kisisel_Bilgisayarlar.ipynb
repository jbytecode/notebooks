## Bilgisayar çalışmaya başladığında ne olur?

Bilgisayar çalışmaya başladığında anakart üzerinde yer alan ve ROM adı verilen bir ünitedeki ufak bir sistem tetiklenir. Bu tetiklenen program oldukça düşük seviyeli bir program olup Assembly veya C gibi dillerde yazılır. Bu program sabit disk bağlantıları, CD rom ve bellek sayımı gibi işlemleri gerçekleştirdikten sonra sabit diskin ilk sektörlerinde yer alan işletim sistemi başlatıcıyı tetikler. Linux işletim sisteminde bu tetikleyici GRUB veya LILO isimler alır (Aynı işi yapan farklı tetikleyicilerdir). Bu düşük seviyeli tetikleyiciler işletim sistemi çekirdeğinin çalıştıralacağı parametreleri tanımlar ve belirler. 

Bir sonraki aşamada işletim sisteminin çalışması sağlanır (Microsoft Windows, Mac OSX veya Linux gibi). İşletim sistemi cihaz sürülerini kurmak, temel girdi çıktı (IO) ünitelerini bağlamak, grafik kartına bağlı olarak masaüstü ortamını hazırlamakla yükümlüdür. İşletim sistemi ayrıca ağ cihazlarına (Ethernet kartı ve wireless) uygun sürücü programlarını (driver) hazırlayıp aradaki iletişim protokollerini kurar. İletişim protokolleri (TCP/IP gibi) daha sonra internet tarayıcı da dahil olmak üzere alt seviye gereksinimleri belirler. 

İşletim sistemi ayrıca kullanıcı tarafından başlatılan programların (internet tarayıcı, kelime işlem veya elektronik hesap tabloları) üzerinde çalışacağı bellek alan ve miktarlarını da ayarlamaktan sorumludur. Kapatılan programlardan ayrılan ve boşa düşen belleğin tekrardan yeni başlatılan programlar için kullanılması, aynı anda çalışan programlar arasında işlemci sürelerinin paylaştırılması gibi bir çok hesap gerektiren işlemler birlikte düşünüldüğünde bilgisayar kaynaklarını tamamen kendine tahsis etmekle etkin çalışmak arasında birçok optimizasyon problemi çözerler.


## Bellek ölçü birimleri

### BIT (BInary digiT) 

3 ve 33 sayıları bilgisayar belleğinde eşit yer mi kaplar? Ya da 'Arda' ile 'Ardahan' bellekte ne kadar yer kaplar? Yer kaplamak ne demektir? 

Kişisel elektronik bilgisayarlar için doğal sayı tabanı 2'dir (Biz insanlar için bu sayı 10'dur). 10'luk sayı sisteminden bahsederken 0, 1, 2, 4, 5, 6, 7, 8 ve 9 rakamlarından oluşan sayılardan bahsediyoruz. Tam tersine 8'lik sayı sistemini kullanan akıllı bir varlık yeryüzünde bizimle birlikte yaşasaydı rakamları 0, 1, 2, 4, 5, 6 ve 7 olan bir sistem kullanacaktı. Ancak gerçekten de bu kadar kalabalığa yer yoktur. 2'li sayı sistemi tüm tamsayıları ifade etmeye yeter.

2'li sayı sisteminde 0 ve 1 rakamları yer alır. Şu ifadeleri inceleyelim:



| Bit (Binary Digit) | Decimal (Onluk Taban Gösterimi) | Nasıl Hesaplanır ? (Modüler aritmatik) |
| :-----: | :-----: | :---- |
| 0  |  0 | $0 \times 2^0$ |
| 1  |  1 | $1 \times 2^0$ |
| 10 |  2 | $1 \times 2^1 + 0 \times 2^0$ |
| 11 |  3 | $1 \times 2^1 + 1 \times 2^0$ | 
| 100|  4 | $1 \times 2^2 + 0 \times 2^1 + 0 \times 2^0$ |
| 101|  5 | $1 \times 2^2 + 0 \times 2^1 + 1 \times 2^0$ |
| 110|  6 | $1 \times 2^2 + 1 \times 2^1 + 0 \times 2^0$ |
| 111|  7 | $1 \times 2^2 + 1 \times 2^1 + 1 \times 2^0$ |
| 1000|  8 | $1 \times 2^3 + 0 \times 2^2 + 0 \times 2^1 + 0 \times 2^0$ |

Dikkat edildiğinde görüleceği üzere, sonu 0 ile biten sayılar çift, 1 ile biten sayılar ise tek olmaktadır. 


### Soru
3 bit ile yazılabilecek en büyük sayının onluk tabandaki karşılığı 7'dir. Bunu tablodan görmekteyiz. Benzer şekilde 4 ile 16, 5 bit ile 32 farklı sayı yazılabilir. Sayı sistemine sıfırın da dahil olduğunu hatırlarsak, örneğin 4 bit ile yazılabilecek tüm sayılar 

$$
0, 1, 2, 3, \dots, 15
$$

olur. O halde diyebiliriz ki $n$ bit ile yazılabilecek tam sayıların sayısı

$$
2^n
$$

olur. 

### Soru
Bir önceki soruyu tersten soralım. 8 sayısını yazabilmek için kaç bit'e ihtiyaç vardır? Biliyoruz ki 3 bit ile yazılabilecek en büyük sayı 7 idi. O halde 8 sayısı için 4 bit gerekir. Tabi ki bunun yanıtını

$$
(1000)_2 = (8)_{10}
$$

eşitliğini sağlayan moduler aritmatik bilgilerimizden hatırlıyoruz.

Aşağıdaki Python kodu, 8 sayısının ikili gösterimini vermektedir. Siz de aynı kodu **sayi = 8** kısmını değiştirerek başka sayılar için deneme yapabilirsiniz.

```python
sayi = 8
bits = []

while True:
    kalan = sayi % 2
    bolum = sayi // 2
    
    bits.append(kalan)
    
    sayi = bolum
    
    if bolum == 0:
        break

    
print(list(reversed(bits)))
```

### Byte

Her ne kadar işlemler bit düzeyinde halledilse de bellek üzerinde tek seferde erişebilir en ufak birim *byte* 'tır (İşin tuhafı hem bit hem de byte İngilizce'de bayt diye okunur ancak biz Türkçe konuşanlar olarak bu karmaşayı yaşamıyoruz). Bir Byte'lık bellek alanı 8 Bit içerir (Dikkat 8 bit içerdiğine göre en fazla $2^8$ adet tam sayı gösterilebilir. Bu sayıyı hesaplamaya hiç ihtiyaç duydunuz mu?

```python
2 ** 8
```

Aşağıda bazı bellek metrikleri arasındaki ilişkiler gösterilmektedir:


| Ölçü Birimi | Kısaltma |Karşılığı Nedir? |
| :-----:     |    :---: | :----: |
|  1 Byte     |   Byte   |   8 Bit |
|  1 Kilobyte |   KB     | 1024 Byte |
|  1 Megabyte |   MB     | 1024 Kilobyte |
|  1 Gigabyte |   GB     | 1024 Megabyte |
|  1 Terabyte |   TB     | 1024 Gigabyte |
|  1 Petabyte |   PB     | 1024 Terabyte |

Son yıllarda $1024$ olan değerler $1000$ olarak yuvarlanıp hesaplamaların kolaylaştırılması sağlanmıştır. Ancak 

$$
2^n = 1000
$$

eşitliğini sağlayan bir $n \in \mathbb{Z}$ olmadığı açıktır. O yüzden teknik konularda $1024$ değerleri geçerlidir. 
Ayrıca 

$$
1024_{10} = (10000000000)_2
$$

olduğunu göstermek faydalı bir alıştırma olabilir.
