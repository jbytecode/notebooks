Yukarıdan bırakılan toplar metal uzantılara çarparak $\frac{1}{2}$ olasılık ile sağa veya sola düşmektedir. Bir sonraki aşamada tekrar bir metal uzantıyla karşılaşan top yine aynı ve eşik olasılıkla sağ veya sola düşer. Bir süre sonra tüm metal uzantılar biter ve top aşağıda bulunan haznelerden birine düşer. Binom dağılımı adı verilen özel bir olasılık dağılımı gereği, bir topun her aşamada sağa veya sola yönelme durumu mümkündür ancak düşük olasılıklıdır. Bu bağlamda toplar ilk bırakıldıkları bölge civarında birikme ve kuyruklarda azalma eğilimi gösterirler. Binomdan normal dağılıma yaklaşım adıyla anılan bu fenomen istatistik disiplininde önemli bir yere sahiptir. Basitlik açısından sıfır ortalamalı ve bir standart sapmalı normal dağılımı yazalım:

$$
f(x; \mu = 0, \sigma^2 = 1) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}x^2}
$$

Burada $-\infty < x < \infty$ olabilir.

```r
# R code
curve(dnorm, from = -4, to = 4)
```

Şimdi şöyle bir oyun düşünelim. $t_0$ anında $x=0$ olsun. $t_1$ anında $x$'in değeri ya $-1$ ya da $1$ olsun. Yani ya 1 birim ufalsın ya da 1 birim büyüsün. Buna $x_1$ diyelim. Sonra $t_2$ anında $x$'in değeri tekrardan ya 1 birim ufalsın ya da 1 birim artsın. $t_{10}$ anında ne olduğunu bir yere kaydedelim: 

```r
set.seed(123)

x <- 0

for (i in 1:10){
    if (runif(1) < 0.5){
        x <- x - 1
    } else{
        x <- x + 1
    }
}

print(x)
```

Şimdi bu deneyi 20 kez tekrarlayıp sonuçları yazdıralım:

```r
for (deney in 1:20) {
    x <- 0

    for (i in 1:10){
        if (runif(1) < 0.5){
            x <- x - 1
        } else{
            x <- x + 1
        }
    }

    print(x)
}
```

Görüldüğü gibi ender olarak 0 civarında değerlere ulaşılmış, yalnızca bir kez -8 değerine ulaşılmış ve 0'dan uzak değerlere pek ulaşılmamış olduğu görülür. Deneyi 10000 kez tekrarlayıp bu kez histogram grafiğini inceleyelim:

```r
sonuclar <- rep(0, 10000)

for (deney in 1:10000) {
    x <- 0

    for (i in 1:10){
        if (runif(1) < 0.5){
            x <- x - 1
        } else{
            x <- x + 1
        }
    }

    sonuclar[deney] <- x
}

hist(sonuclar)
head(sonuclar)
```
