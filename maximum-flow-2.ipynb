```julia
using JuMP, HiGHS
```

![Maximum Flow Example 2](maximum-flow-2.png)

```julia
begin
    m = Model(HiGHS.Optimizer)
    @variable(m, f >= 0)
    @variable(m, x12 >= 0)
    @variable(m, x13 >= 0)
    @variable(m, x34 >= 0)
    @variable(m, x25 >= 0)
    @variable(m, x56 >= 0)
    @variable(m, x46 >= 0)
    @objective(m, Max, f)
    @constraint(m, x12 + x13 == f)
    @constraint(m, x46 + x56 == f)
    @constraint(m, x12 == x25)
    @constraint(m, x34 == x46)
    @constraint(m, x25 == x56)
    @constraint(m, x13 <= 15)
    @constraint(m, x12 <= 10)
    @constraint(m, x25 <= 25)
    @constraint(m, x56 <= 12)
    @constraint(m, x34 <= 20)
    @constraint(m, x46 <= 13)
    optimize!(m)
    println("f = ", value(f))
    println(value.([x12, x13, x34, x25, x56, x46, f]))
    println(m)
end
```
