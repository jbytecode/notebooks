# Diet Problem with Two-Stage Simplex

|           | Medicine 1    | Medicine 2   | Requirement   |
| :---:     | :----------:  | :---------:  | :---------:   |
| Vitamin A |     3         |      4       |     120       |
| Vitamin B |     5         |      3       |     150       |
| Price     |    10         |      12      |               |

$$
\begin{aligned}
\min z = & 10x + 12y \\
\text{Subject to:} & \\
& 3x + 4y \ge 120 \\
& 5x + 3y \ge 150 \\
& x, y \ge 0 
\end{aligned}
$$

$$
\begin{aligned}
\min z = & 10x + 12y \\
\text{Subject to:} & \\
& 3x + 4y - s_1 + R_1 = 120 \\
& 5x + 3y - s_2 + R_2 = 150 \\
& x, y \ge 0 
\end{aligned}
$$

$$
R_1 = 120 - 3x - 4y + s_1
$$

$$
R_2 = 150 - 5x - 3y + s_2 
$$

- Let's minimize total R values, that is, $\min W = \Sigma_{i=1}^2 R_i$
- if total $R_i$ values are zero, then the solution if feasible

$$
R_1 + R_2 = 270 - 8x - 7y + s_1 + s_2
$$

|   V    |    $x_1$   |   $x_2$    |   $s_1$   |  $s_2$  |  $R_1$  | $R_2$  | Solution   |
| :----:  |  :------:  | :------:   | :-------: | :-----: | :----:  | :----: | :--------: |
| $W$   |      8     |     7      |    -1     |    -1   |    0    |   0    |  270       |
| $R_1$ |      3     |     4      |    -1     |     0   |    1    |   0    |  120       |
| $R_2$ |      5     |     3      |    0      |    -1   |    0    |   1    |  150       |

- Enterin variable is $x_1$
- Exiting variable is $R_2$

```julia
x1 = Rational[5, 3, 0, -1, 0, 1, 150] / Rational(5)
```

```julia
R1 = Rational[3, 4, -1, 0, 1, 0, 120];
R1 = R1 + x1 * (-3)
```

```julia
W = Rational[8, 7, -1, -1, 0, 0, 270];
W = W + x1 * (-8)
```

|       |    $x_1$   |   $x_2$    |   $s_1$   |  $s_2$  |  $R_1$  | $R_2$  | Solution   |
| :--:  |  :------:  | :------:   | :-------: | :-----: | :----:  | :----: | :--------: |
|  $W$  |      0     |   11/5     |    -1     |    3/5  |    0    |  -8//5 |  30       |

| $x_1$ |      1     |   3/5      |    0      |   -1/5  |    0    |   1/5  |  30       |

- $x_2$ is entering variable
- $R_1$ is exiting variable

```julia
R1 = R1 / (11//5)
```

```julia
x2 = R1;
```

```julia
x1 = x1 + x2 * (-3//5)
```

```julia
W = W + x2 * (-11//5)
```

|   V    |    $x_1$  |   $x_2$    |   $s_1$   |  $s_2$  |  $R_1$  | $R_2$  | Solution   |
| :----: |  :------: | :------:   | :-------: | :-----: | :----:  | :----: | :--------: |
|  $W$  |      0     |   0        |    0      |   0     |   -1    |  -1    |  0       |

- No entering variable
- Optimal $W$ is zero, so the solution is feasible.
- $x_1$ is $240/11$, $x_2$ is $150/11$
- $x_2 - \frac{5}{11}s_1 + \frac{3}{11}s_2 = \frac{150}{11}$
- $x_1 + \frac{3}{11}s_1 - \frac{4}{11}s_2 = \frac{240}{11}$
- $z = 10x_1 + 12x_2$
- $z = 10\Big[\frac{240}{11} - \frac{3}{11}s_1 + \frac{4}{11}s_2\Big] + 12\Big[\frac{150}{11} + \frac{5}{11}s_1 - \frac{3}{11}s_2 \Big]$
- $z = \frac{4200}{11} + \frac{30}{11}s_1 + \frac{4}{11}s_2$

## JuMP and HiGHS Solution

```julia
using Pkg; Pkg.activate(".")
```

```julia
using JuMP, HiGHS
```

```julia
begin 
    m = Model(HiGHS.Optimizer)
    @variable(m, x>=0)
    @variable(m, y>=0)
    @objective(m, Min, 10x + 12y)
    @constraint(m, 3x + 4y >= 120)
    @constraint(m, 5x + 3y >= 150)
    optimize!(m)
end 
```

```julia
value(x)
```

```julia
value(y)
```

```julia
objective_value(m)
```
