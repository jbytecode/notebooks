```julia
using OperationsResearchModels;
```

```julia
costs = [
    8 13 25 25 0;
    15 12 26 25 0;
    0 6 16 17 0;
    6 0 14 16 0
];
```

```julia
demand = [350, 350, 130, 130, 90];
```

```julia
supply = [150, 200, 350, 350];
```

```julia
problem = TransportationProblem(costs, demand, supply);
```

```julia
result = solve(problem);
```

```julia
result.cost
```

```julia
result.solution
```
