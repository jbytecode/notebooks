```python
# Our first program. Calculate roots of a parabola
# The parabola is a function in form of 
# f(x) = ax^2 + bx + c
# where a, b, c are constants and x is the independent variable.
# delta = b^2 - 4ac
# if delta > 0, there are two distinct roots.
# if delta < 0, there is no any real roots.
# x1 = (-b + sqrt(delta)) / 2a 
# and 
# x2 = (-b - sqrt(delta)) / 2a

a = -1
b = 10
c = 15
delta = b * b - 4 * a * c
x1 = (-b + delta ** (-0.5)) / (2 * a)
x2 = (-b - delta ** (-0.5)) / (2 * a)
print(delta)
print(x1)
print(x2)
```

```python
# Note:
# a ** b is a^b, where a is the base and b is the power.
```
