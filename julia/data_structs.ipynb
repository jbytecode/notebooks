```julia
struct Language
    name::AbstractString
    age::Int
end
```

```julia
p1 = Language("Julia", 10)
```

```julia
p2 = Language("Python", 20)
```

```julia
p3 = Language("Java", 26)
```

```julia
p4 = Language("C", 45)
```

```julia
Base.isless(lang1::Language, lang2::Language)::Bool = lang1.age < lang2.age
```

```julia
languages = [p4,p2, p1, p3]
```

```julia
sort(languages)
```

```julia
Base.:+(lang1::Language, lang2::Language) = Language(lang1.name * " and " * lang2.name, lang1.age + lang2.age)
```
