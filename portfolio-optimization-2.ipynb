```julia
using JuMP, Ipopt, Statistics, Random
```

```julia
mt = MersenneTwister(1234)
```

```julia
data = rand(mt, 5, 2)
```

```julia
m = mean(data, dims = 1)
```

```julia
v = cov(data)
```

```julia
begin
    model = Model(Ipopt.Optimizer)
    @variable(model, w[1:2] >= 0)
    @constraint(model, sum(w) == 1.0)
    @constraint(model, sum(w[i] * m[i] for i in 1:2) >= 0.45)
    @objective(model, Min, w'*v*w)
    optimize!(model)
end
```

```julia
JuMP.is_solved_and_feasible(model)
```

```julia
weights = value.(w)
```

```julia
JuMP.objective_value(model)
```

```julia
weights' .* m |> sum
```

```julia
JuMP.latex_formulation(model)
```
