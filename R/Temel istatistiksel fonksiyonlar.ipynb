```r
# Elimizdeki tek değişkenli veri aşağıdaki gibi olsun:
my.data <- c(1, 6, 7, -5, 9, 10, 14, 21, 77, 43, 21, 21, 97, 3, 45)
```

```r
my.data
```

```r
# Özet istatistikleri elde edelim:
# Ortalama
mean(my.data)
```

```r
# Medyan
median(my.data)
```

```r
# Varyans
var(my.data)
```

```r
# Standart sapma
sd(my.data)
```

```r
# Standart sapma (veya)
sqrt(var(my.data))
```

```r
# Covaryans
cov(my.data, my.data)
```

```r
cov(my.data, my.data) == var(my.data)
```

```r
# x ve y Uniform(0,1) dağılımından rastgele elde edilsin:
x <- runif(n = 15, min = 0, max = 1)
y <- runif(n = 15, min = 0, max = 1)
```

```r
cbind(x, y)
```

```r
# Kovaryans ve Korelasyon
cov(x, y)
```

```r
cor(x, y)
```

```r
cor(x, y, method = "?")
```

```r
# Görüldüğü gibi Pearson, Kendall ve Spearman korelasyonları hesaplanabilir:
cor(x, y, method = "pearson")
```

```r
cor(x, y, method = "spearman")
```

```r
# veri adlı rassal çekilen veri için Normallik testi uygulayalım:
# tseries paketi yoksa, install.packages() ile önceden kurulmalı.
set.seed(123)
x <- runif(n = 200, min = 0, max = 1)
library(tseries)
```

```r
jarque.bera.test(x)
```

```r
# p-value: 0.004786, verinin normal dağılım sergilemediği söylenebilir. 
# Bunu bir histogram ile destekleyelim:
hist(x, col = "lightgreen")
```

```r
# Normal dağılımdan rasgele çekilmiş değerler için sonuçlar elbette farklı olacaktı:
# N(0, 1) dağılımından 1000 adet rassal sayı oluşturalım:
set.seed(123)
z <- rnorm(1000, mean = 0, sd = 1)
hist(z, col = "lightblue")
```

```r
jarque.bera.test(z)
```

```r
# z'nin ortalaması
mean(z)
```

```r
# Tabi ki bu ortalamanın örnekte elde edildiğini biliyoruz.
# Acaba bu örnek ortalaması 10 olan bir kitleden (population)
# çekilmiş olabilir mi?
t.test(z, mu = 10.0, paired = FALSE)
```
