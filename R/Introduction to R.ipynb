```r
# Bu bir açıklama satırı ve işletilmez.
# Burada a = -1, b = 5, c = 10 olan ax^2 + bx + c parabolunun
# kökleri aranmaktadır
a <- -1
b <- 5
c <- 10
delta <- b^2 - 4 * a * c
x1 <- (-b + sqrt(delta)) / (2 * a)
x2 <- (-b + sqrt(delta)) / (2 * a)
print (c(x1, x2)) 
```
