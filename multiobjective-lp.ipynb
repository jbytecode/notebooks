## A Problem

|             |      A        |       B       |    Capacity       |
| :--------:  | :---------:   | :----------:  | :------------:    |
| Labor       |       3       |    2          |  120              |
| Machine     |       2       |    4          |     240           |
| Unit Profit |        5      |    6          |                   | 
| Unit Cost   |        3      |    2          |                   |

$$
\begin{aligned}
\max z = & 5x_1 + 6x_2 \\
\min w = & 3x_1 + 2x_2 \\
\text{Subject to:} & \\
& 3x_1 + 2x_2 \le 120 \\
& 2x_1 + 4x_2 \le 240 \\
& x_1, x_2 \ge 0 \\
\end{aligned}    
$$

![Multi-objective-lp](multiobjective-lp.png)

|    Point     |      z       |      w      |    z-w      |
| :---------:  | :---------:  | :--------:  | :-------:   |
|  (0,0)       |      0       |      0      |    0        |
|  (40,0)      |     200      |      120    |   80        |
|  (0,60)      |     360      |      120    |  240        |

- In point $(0,60)$, $z$ has the maximum value.
- In point $(0,60)$, $w# has the minimum value (among two points for 120)

Now, define the composite profit function:

$$
\begin{aligned}
\max T\pi = & 5x_1 + 6x_2 - [3x_1 + 2x_2] = 2x_1 + 5x_2 \\
\text{Subject to:} & \\
& 3x_1 + 2x_2 \le 120 \\
& 2x_1 + 4x_2 \le 240 \\
& x_1, x_2 \ge 0 \\
\end{aligned}    
$$

|      Point     |  Total Profit  |
|  :--------:    | :----------:   |
|  (0,0)         |   0            |
|  (40, 0)       |  80            |
|  (0,60)        |  240           |
